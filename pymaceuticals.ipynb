```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy.stats import linregress

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data 
merge = pd.merge(mouse_metadata, study_results, on="Mouse ID", how="left")
merge.head()
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.
mean = merge.groupby('Drug Regimen')['Tumor Volume (mm3)'].mean()
median = merge.groupby('Drug Regimen')['Tumor Volume (mm3)'].median()
variance = merge.groupby('Drug Regimen')['Tumor Volume (mm3)'].var()
stdv = merge.groupby('Drug Regimen')['Tumor Volume (mm3)'].std()
sem = merge.groupby('Drug Regimen')['Tumor Volume (mm3)'].sem()
summary_df = pd.DataFrame({"Mean": mean, 
                           "Median": median, 
                           "Variance": variance,
                           "Standard Deviation": stdv, 
                          "SEM": sem})

summary_df
```

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
drug_df=merge[merge['Drug Regimen'].isin(['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin'])]

# Start by getting the last (greatest) timepoint for each mouse

last_df=drug_df.groupby(['Mouse ID']).agg({'Timepoint':['max']})
last_df.columns=['Max Timepoint']
# Merge this group df with the original dataframe to get the tumor volume at the last timepoint

output_df=pd.merge(last_df, drug_df, how='left',on=['Mouse ID'])

output_df['diff']=output_df['Max Timepoint'] - output_df['Timepoint']
output_df
```

```python
#Put treatments into a list for for loop (and later for plot labels)
# Create empty list to fill with tumor vol data (for plotting)
plot_list=[]
# Calculate the IQR and quantitatively determine if there are any potential outliers. 
    # Locate the rows which contain mice on each drug and get the tumor volumes
    # add subset 
    # Determine outliers using upper and lower bounds
    
lp_vol=output_df.loc[(output_df['diff'] == 0) & (output_df['Max Timepoint']> 0)]
print(lp_vol)
lp_vol=lp_vol[['Mouse ID', 
               'Timepoint', 
               'Tumor Volume (mm3)', 
               'Drug Regimen',
               'Weight (g)']]
        # Get all drugs
drug_list = lp_vol['Drug Regimen'].unique()
        # df.tolist() converts dataframe to list
drug_list=drug_list.tolist()

    


        #for loops for all drugs in Drug_list
for drug_regimen in drug_list:
    drug_group=lp_vol.loc[lp_vol['Drug Regimen']==drug_regimen]
    tumor_vol=drug_group['Tumor Volume (mm3)']
 
   
    plot_list.append(tumor_vol)    

    quartiles = tumor_vol.quantile([.25, .5, .75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr=round(upperq-lowerq,2)
    print(f'IQR for '+str(drug_regimen)+' is '+ str(iqr))
    
    
    lower_b=round(lowerq-(1.5*iqr),2)
    upper_b=round(upperq+(1.5*iqr),2)
    
    print(f'Lower Bound for' + str(drug_regimen)+ ' : ' + str(lower_b))
   
    print(f'Upper Bound for'+ str(drug_regimen) + ' : ' + str(upper_b))
    
    
    outliers=(tumor_vol.loc[(lp_vol['Tumor Volume (mm3)'] >= upper_b) | (lp_vol['Tumor Volume (mm3)'] <= lower_b)]).count()
    
    print(f'The number of outliers for '+str(drug_regimen)+ ' is '+ str(outliers)) 
    print(f'... ')
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
# Generate a box plot of the final tumor volume of each mouse
plt.boxplot(plot_list)
plt.title('Tumor Volume Data by Regimen')
plt.ylabel('Tumor Volume (mm3)')
drug=[1,2,3,4]
plt.xticks(drug, drug_list)
plt.show()
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen

capomulin_df = merge.loc[merge["Drug Regimen"] == "Capomulin"]
capomulin_df = capomulin_df.reset_index()
capomulin_df.head()

cap_single_df = capomulin_df.loc[capomulin_df["Mouse ID"] == "k403"]

reduced_df = capomulin_df.loc[:, ["Mouse ID", "Weight (g)", "Tumor Volume (mm3)"]]

avg_cap = pd.DataFrame(reduced_df.groupby(["Mouse ID", "Weight (g)"])["Tumor Volume (mm3)"].mean()).reset_index()

avg_cap = avg_cap.rename(columns={"Tumor Volume (mm3)": "Average Volume"})
avg_cap = avg_cap.set_index('Mouse ID')
avg_cap.plot(kind="scatter", x="Weight (g)", y="Average Volume", grid=True, figsize=(3,3),
              title="Weight Vs. Average Tumor Volume")
plt.show()
```

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen


weight = avg_cap.iloc[:,0]
avg_tumor_volume = avg_cap.iloc[:,1]
coral = st.pearsonr(weight,avg_tumor_volume)
print(f'The correlation between both factors is '+ str(round(coral[0],2)))


x_values = avg_cap['Weight (g)']
y_values = avg_cap['Average Volume']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress = x_values * slope + intercept
line = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
print(str(line))
        #scatter plot
plt.scatter(x_values,y_values)
        #linear regression
plt.plot(x_values,regress,"r-")
plt.xlabel('Mouse Weight')
plt.ylabel('Average Tumor Volume')
plt.show()
```
